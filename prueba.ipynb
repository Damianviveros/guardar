```python
import numpy as np
from matplotlib import pyplot as plt
import pylab as pl
x=np.linspace(-4,4)
for C in range (1,4,1):
    y=C*pow(x,3)
    p=pl.subplot()
    p.plot(x,y,label="PARA C={}".format(C))
p.set_xlim(-5,5)
plt.axhline(0,color="black")
plt.axvline(0,color="black")

p.set_ylabel("y")
p.set_xlabel("x")
p.legend(loc="best")
plt.xticks(range(-5,5,1))
plt.yticks(range(-200,200,10))
plt.rcParams["figure.figsize"]=(10,6)
p.grid()
plt.show()
```
